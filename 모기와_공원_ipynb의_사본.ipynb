## 0-2. pandas 핵심 — 표(DataFrame) 다루기

| 동작 | 함수 |
|------|------|
| 작은 표 만들기 | `pd.DataFrame()` |
| 데이터 살펴보기 | `head()` / `info()` / `shape` / `columns` |
| 새 컬럼 만들기 | `df["새컬럼"] = ...` |
| 조건 필터링 | `df[df["컬럼"] >= 값]` |
| 그룹별 집계 | `groupby().sum()` / `.mean()` |
| 정렬 | `sort_values()` |

| 0단계에서 배운 것 | 본 수업에서 |
|------|------|
| 리스트 · 딕셔너리 · 변수 | 파일 경로·컬럼 변수 만들기 (1단계) |
| `head()` · `info()` · `columns` | 데이터 탐색 EDA (2단계) |
| 새 컬럼 만들기 `df["컬럼명"] = ...` | 총 이용 건수 · 순유출 생성 (3·4단계) |

| `groupby().sum()/.mean()` | 대여소별·자치구별·월별 집계 (3·4단계) |
| `sort_values()` | 순위표·TOP 20 (4·5단계) |
| f-string `f"{값}"` | 결과 문장·종합표 자동완성 (4단계) |

````python
import pandas as pd

# '모기지수 발생일' 컬럼에서 연도 추출
mosquito_df['발생연도'] = mosquito_df['모기지수 발생일'].dt.year

# 2020년을 기준으로 데이터 분리
mosquito_before_2020 = mosquito_df[mosquito_df['발생연도'] < 2020]
mosquito_2020_onwards = mosquito_df[mosquito_df['발생연도'] >= 2020]

# 각 기간별 '모기지수(공원)'의 평균 계산
avg_before_2020 = mosquito_before_2020['모기지수(공원)'].mean()
avg_2020_onwards = mosquito_2020_onwards['모기지수(공원)'].mean()

print(f"2020년 이전 모기지수(공원) 평균: {avg_before_2020:.2f}")
print(f"2020년 이후 모기지수(공원) 평균: {avg_2020_onwards:.2f}")
````

### 새로운 감염병 매개 모기 데이터 불러오기

````python
import glob
import unicodedata

# 원래 패턴 문자열
original_pattern = '2025년 * 감염병 매개모기 채집결과(유문등).xlsx'
# 한글 유니코드 정규화 (NFD: Canonical Decomposition)
normalized_pattern = unicodedata.normalize('NFD', original_pattern)

# 정규화된 패턴을 사용하여 새로운 엑셀 파일 경로 목록 생성
m_mosquito_files = glob.glob(folder_path + normalized_pattern)

# 파일 목록 출력 확인
print("불러올 엑셀 파일 목록:")
for f in m_mosquito_files:
    print(f)

# 각 엑셀 파일을 읽어 리스트에 저장
m_mosquito_dfs = []
for file in m_mosquito_files:
    try:
        df_temp = pd.read_excel(file)
        m_mosquito_dfs.append(df_temp)
    except Exception as e:
        print(f"Error reading {file}: {e}")

# 모든 데이터프레임을 하나로 합치기
if m_mosquito_dfs:
    monthly_mosquito_df = pd.concat(m_mosquito_dfs, ignore_index=True)
    print("\n모든 엑셀 파일 병합 완료!")
    print("\n병합된 데이터프레임 (monthly_mosquito_df) 상위 5행:")
    display(monthly_mosquito_df.head())
    print("\n병합된 데이터프레임 정보:")
    monthly_mosquito_df.info()
else:
    print("불러올 엑셀 파일이 없거나 오류로 인해 데이터프레임을 생성할 수 없습니다.")
````

새로운 모기 데이터(`monthly_mosquito_df`)의 컬럼명이 `Unnamed`로 표시되어 있어 데이터 구조를 파악하기 어렵습니다. 이는 엑셀 파일에 여러 헤더 행이 있기 때문입니다. 올바른 헤더를 읽어와 컬럼명을 정리하고, 데이터 타입을 확인하여 전처리하겠습니다.

````python
import glob
import unicodedata
import pandas as pd
import numpy as np

original_pattern = '2025년 * 감염병 매개모기 채집결과(유문등).xlsx'
normalized_pattern = unicodedata.normalize('NFD', original_pattern)
m_mosquito_files = glob.glob(folder_path + normalized_pattern)

m_mosquito_dfs_cleaned = []

for file in m_mosquito_files:
    try:
        # Header가 2줄(index 1, 2)인 구조를 반영
        df_temp = pd.read_excel(file, header=[1, 2])

        # 파일명에서 월 추출
        filename_nfc = unicodedata.normalize('NFC', file.split('/')[-1])
        month_str = filename_nfc.split('년 ')[1].split('월')[0]

        # 멀티인덱스 컬럼 정리: (Level 0, Level 1) -> 유의미한 이름 선택
        new_cols = []
        for col in df_temp.columns:
            c0, c1 = str(col[0]), str(col[1])
            if 'Unnamed' in c0:
                new_cols.append(c1) # 월별, 채집일 등
            elif 'Unnamed' in c1:
                new_cols.append(c0) # 합계 등
            else:
                new_cols.append(c1) # 모기 종 이름

        df_temp.columns = new_cols
        df_temp['월'] = int(month_str)

        # 첫 행(설명 텍스트) 제거 및 데이터 행만 필터링
        df_temp = df_temp.dropna(subset=['월별']).iloc[1:].copy()
        m_mosquito_dfs_cleaned.append(df_temp)

    except Exception as e:
        print(f"Error reading {file}: {e}")

if m_mosquito_dfs_cleaned:
    monthly_mosquito_df_processed = pd.concat(m_mosquito_dfs_cleaned, ignore_index=True)

    # 불필요한 'Unnamed' 키워드가 남은 컬럼 최종 제거
    cols = [c for c in monthly_mosquito_df_processed.columns if 'Unnamed' not in str(c)]
    monthly_mosquito_df_processed = monthly_mosquito_df_processed[cols]

    # 컬럼 순서 재배치 (월, 월별, 채집일 순)
    first_cols = ['월', '월별', '채집일']
    other_cols = [c for c in monthly_mosquito_df_processed.columns if c not in first_cols]
    monthly_mosquito_df_processed = monthly_mosquito_df_processed[first_cols + other_cols]

    # 숫자형 변환
    for col in other_cols:
        monthly_mosquito_df_processed[col] = pd.to_numeric(monthly_mosquito_df_processed[col], errors='coerce').fillna(0)

    print("2025년 상세 데이터 병합 및 정제 완료!")
    display(monthly_mosquito_df_processed.head())
    monthly_mosquito_df_processed.info()
else:
    print("파일을 찾을 수 없습니다.")
````

### **2026년 맞춤형 방역 전략 적용 시 예상 감소 효과 예측**

2025년의 지역별 피크 패턴이 2026년에도 유지된다고 가정하고, 제안된 **'시공간 차등 방역 로드맵'**을 적용했을 때의 모기지수 감소 시뮬레이션을 수행합니다.

*   **수변부 (Waterfronts):** 5~6월 유충 방제 집중 시 약 35% 감소 예측
*   **주거지 (Residential):** 7~8월 정화조/하수구 집중 관리 시 약 20% 감소 예측
*   **공원 (Parks):** 8~9월 방역 기간 연장 시 약 25% 감소 예측

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 2025년 데이터를 2026년 베이스라인(Base)으로 설정
prediction_2026 = loc_monthly_avg.copy()

# 지역별 예상 감소율 설정 (시나리오 기반)
reduction_rates = {
    '모기지수(수변부)': 0.35,  # 35% 감소
    '모기지수(주거지)': 0.20,  # 20% 감소
    '모기지수(공원)': 0.25     # 25% 감소
}

# 맞춤형 방역 적용 후 예상 지수 계산
for col, rate in reduction_rates.items():
    prediction_2026[f'{col}_예상'] = prediction_2026[col] * (1 - rate)

# 시각화
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
colors = ['blue', 'green', 'red']
labels = ['수변부', '주거지', '공원']

for i, col in enumerate(loc_monthly_avg.columns):
    axes[i].plot(prediction_2026.index, prediction_2026[col], marker='o', linestyle='--', color='gray', label='2025년 (기존)')
    axes[i].plot(prediction_2026.index, prediction_2026[f'{col}_예상'], marker='s', linestyle='-', color=colors[i], label='2026년 (예상)')
    axes[i].set_title(f'{labels[i]} 방역 효과 예측')
    axes[i].set_xlabel('월')
    axes[i].set_ylabel('평균 모기지수')
    axes[i].legend()
    axes[i].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# 예상 감소 효과 요약
print("=== 2026년 맞춤형 방역 전략 적용 시 기대 효과 (연간 평균 지수 기준) ===")
for col in loc_monthly_avg.columns:
    base = prediction_2026[col].mean()
    pred = prediction_2026[f'{col}_예상'].mean()
    diff = base - pred
    print(f"[{col.replace('모기지수(', '').replace(')', '')}] 예상 평균 감소폭: {diff:.2f} p ({ (diff/base)*100:.1f}% 감소)")
````

### **[추가 분석] 방역 자원 최적 배분 모델 및 기후 변동성 시나리오**

이 분석에서는 앞서 도출된 지역별 피크 시기를 바탕으로 예산과 인력을 어떻게 배분해야 효율적인지 모델링하고, 기후 변화(기온 상승)에 따른 리스크를 점검합니다.

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. 방역 자원 배분 최적화 모델링 (월별 가중치 기반)
# 지역별 피크 강도를 기반으로 필요 방역 강도(Resource Score) 계산
resource_model = loc_monthly_avg.copy()

# 가중치: 수변부(유충방제 중요도) 1.2, 주거지(인구 밀집도) 1.5, 공원 1.0
resource_model['필요_자원량'] = (
    resource_model['모기지수(수변부)'] * 1.2 +
    resource_model['모기지수(주거지)'] * 1.5 +
    resource_model['모기지수(공원)'] * 1.0
)

# 전체 예산 대비 월별 배분 비율 (%) 계산
total_score = resource_model['필요_자원량'].sum()
resource_model['예산_배분율(%)'] = (resource_model['필요_자원량'] / total_score) * 100

# 시각화: 월별 권장 자원 투입 비중
plt.figure(figsize=(10, 5))
plt.bar(resource_model.index, resource_model['예산_배분율(%)'], color='teal', alpha=0.7)
plt.title('2026년 효율적 방역 자원(예산/인력) 월별 배분 제안', fontsize=14)
plt.xlabel('월')
plt.ylabel('배분 비중 (%)')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()

print("=== 최적 자원 배분 제안 (월별) ===")
display(resource_model[['예산_배분율(%)']].round(1).T)

# 2. 기후 변동성 보정 시나리오 (기온 1.5도 상승 시)
# 온도 상승에 따른 모기 활동성 증가 계수 (가정: 1도당 지수 10% 증가)
temp_rise_factor = 1.15
resource_model['기후변화_리스크_지수'] = resource_model['필요_자원량'] * temp_rise_factor

print(f"\n* 기후 변화(기온 1.5도 상승) 고려 시 전체 방역 부하가 약 { (temp_rise_factor-1)*100 } % 증가할 것으로 예상됩니다.")
````
